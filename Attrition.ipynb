## Factors of Attrition

#### 1. Age

```python
sns.distplot(df[df["Attrition"]=="Yes"]["Age"],bins=20,hist=False,kde_kws={"color": "darkslateblue", "fill": True},label="Yes")
sns.distplot(df[df["Attrition"]=="No"]["Age"],bins=20,hist=False,kde_kws={"color": "navajowhite", "fill": True},label="No")
plt.legend()
plt.title("Age wise Attrition")
plt.show()
```

##### The younger age groups are more likely to leave the company compared to their older counterparts.Older employees might have more settled family lives, making them less likely to relocate or make major career changes.



#### 2. Business Travel

```python
temptable1=df.pivot_table(index="BusinessTravel",columns="Attrition",values="EmployeeCount",aggfunc="count")
```

```python
fig, axs=plt.subplots()
plt.suptitle("BusinessTravel wise Attrition")
axs.bar(range(1,df["BusinessTravel"].nunique()+1),list(temptable1["No"]),label="No",color="navajowhite")
axs.bar(range(1,df["BusinessTravel"].nunique()+1),list(temptable1["Yes"]),bottom =list(temptable1["No"]), label="Yes",color="darkslateblue")
axs.set_xticks(range(1,df["BusinessTravel"].nunique()+1),list(df["BusinessTravel"].unique()))
axs.set_xlabel("BusinessTravel")
axs.set_ylabel("Count of Employees")
for bar in axs.patches:
  axs.text(bar.get_x() + bar.get_width() / 2,
          bar.get_height() / 2 + bar.get_y(),
          round(bar.get_height()), ha = 'center',
          color = 'black', weight="bold", size = 9)
axs.legend()
plt.show()
```

```python
print("Attrition Rates (%) on basis of BusinessTravel")
pd. DataFrame(round(df.groupby("BusinessTravel")["Attrition"].value_counts(normalize=True)*100,0))
```

##### Employees who have to travel frequently have a higher attrition rate .It would be essential for the company to include review of travel policies and provide support to employees who travel frequently.

```python
print("             Who are travelling frequently?")
df.pivot_table(index="JobRole",columns="BusinessTravel",values="EmployeeCount",aggfunc="count")
```

##### Most of the employee who have to travel frequently  are Laboratory Technician, Research Scientists and Sales Executive 

#### 3. Department

```python
temptable=df.pivot_table(index="Department",columns="Attrition",values="EmployeeCount",aggfunc="count")
```

```python
fig, axs=plt.subplots()
plt.suptitle("Department wise Attrition")
axs.bar(range(1,df["Department"].nunique()+1),list(temptable["No"]),label="No",color="navajowhite")
axs.bar(range(1,df["Department"].nunique()+1),list(temptable["Yes"]),bottom =list(temptable["No"]), label="Yes",color="darkslateblue")
axs.set_xticks(range(1,df["Department"].nunique()+1),list(df["Department"].unique()))
axs.set_xlabel("Departments")
axs.set_ylabel("Count of Employees")
for bar in axs.patches:
  axs.text(bar.get_x() + bar.get_width() / 2,
          bar.get_height() / 2 + bar.get_y(),
          round(bar.get_height()), ha = 'center',
          color = 'black', weight="bold", size = 9)
axs.legend()
plt.show()
```

```python
print("Attrition Rates in %")
pd. DataFrame(round(df.groupby("Department")["Attrition"].value_counts(normalize=True)*100,0))
```

##### The attrition rates for the Sales and Human Resources departments are higher than that of the Research & Development department.


#### 4. EducationField

```python
df[df["Attrition"]=="Yes"]["EducationField"].value_counts().plot(kind="bar",color= "darkslateblue")
```

##### Employees from Life Sciences and Medical Field are having higher attrition rate. These fields are often focused on researching new theories, applications, and products, making them highly relevant to demanding positions and better career opportunities.

#### 5. Job Role

```python
df.pivot_table(index="JobRole",columns="Attrition",values="EmployeeCount",aggfunc="count").plot(kind="barh",stacked=True,color=["navajowhite","darkslateblue"])
plt.xlabel("Attrition")
plt.title("JobRole wise Attrition")
plt.show()
```

```python
temptable4=df.pivot_table(index="JobRole",columns="Attrition",values="EmployeeCount",aggfunc="count")
```

```python
print("Attrition Rate in different Job Roles")
pd. DataFrame(round((temptable4["Yes"]/(temptable4["Yes"]+temptable4["No"]))*100,0))
```

##### The attrition rate is higher in the "Sales Representative" job role, where it stands at 40%.Following the "Sales Representative" job role, the next highest attrition rates are observed in the "Human Resources" and "Laboratory Technician" roles, with attrition rates of 23% and 24% respectively.


#### 6. Gender

```python
fig,axs=plt.subplots(ncols=2)
axs[0].set_title("Female")
df[df["Gender"]=="Female"]["Attrition"].value_counts().plot(kind="pie",autopct="%.0f%%",ax=axs[0],colors=["navajowhite","darkslateblue"])
axs[1].set_title("Male")
df[df["Gender"]=="Male"]["Attrition"].value_counts().plot(kind="pie",autopct="%.0f%%",ax=axs[1],colors=["navajowhite","darkslateblue"])
plt.suptitle("Gender wise Attrition")
```

##### Male and Female have the same attrition rate. Hence Gender is not a factor of Attrition

#### 7.Marital Status

```python
plt.figure(figsize=(20,7))
fig,axs=plt.subplots(ncols=3)
axs[0].set_title("Single")
df[df["MaritalStatus"]=="Single"]["Attrition"].value_counts().plot(kind="pie",autopct="%.0f%%",ax=axs[0],colors=["navajowhite","darkslateblue"])
axs[1].set_title("Married")
df[df["MaritalStatus"]=="Married"]["Attrition"].value_counts().plot(kind="pie",autopct="%.0f%%",ax=axs[1],colors=["navajowhite","darkslateblue"])
axs[2].set_title("Divorced")
df[df["MaritalStatus"]=="Divorced"]["Attrition"].value_counts().plot(kind="pie",autopct="%.0f%%",ax=axs[2],colors=["navajowhite","darkslateblue"])
```

##### The analysis suggests that single employees have a significantly higher attrition rate compared to employees who are married or divorced

#### 9. Monthly Income

```python
sns.distplot(df[df["Attrition"]=="Yes"]["MonthlyIncome"],kde_kws={"color": "darkslateblue", "fill": True},label="Yes",hist=False)
sns.distplot(df[df["Attrition"]=="No"]["MonthlyIncome"],kde_kws={"color": "navajowhite", "fill": True},label="No",hist=False)
plt.legend()
```

##### Employees with Monthly Income less than 5000 are more prone to leave the company 

#### 10. Distance From Home

```python
sns.distplot(df[df["Attrition"]=="Yes"]["DistanceFromHome"],kde_kws={"color": "darkslateblue", "fill": True},label="Yes",hist=False)
sns.distplot(df[df["Attrition"]=="No"]["DistanceFromHome"],kde_kws={"color": "navajowhite", "fill": True},label="No",hist=False)
plt.legend()
```

##### As the distance of the home from the workplace increases attrition rate also increases.It could be beneficial for companies to consider strategies that could help employees manage their commutes better, such as flexible working hours to avoid peak traffic, work-from-home policies, or even providing transportation allowances.

#### 11.Experience with the company

```python
sns.distplot(df[df["Attrition"]=="Yes"]["YearsAtCompany"],kde_kws={"color": "darkslateblue", "fill": True},label="Yes",hist=False)
sns.distplot(df[df["Attrition"]=="No"]["YearsAtCompany"],kde_kws={"color": "navajowhite", "fill": True},label="No",hist=False)
plt.legend()
plt.title("Experience and Attrition")
```

##### It is evident that the new employees are having more attrition rate"
